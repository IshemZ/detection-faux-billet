```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
billets = pd.read_csv('../Source/billets.csv', delimiter=';')
```

```python
billets
```

```python
billets.describe()
```

```python
billets.info()
```

```python
missing_values = billets[billets['margin_low'].isnull()]
complete_values = billets[billets['margin_low'].notnull()]
```

```python
features = ['diagonal', 'height_left', 'height_right', 'margin_up', 'length']
```

```python
X_train = complete_values[features]
y_train = complete_values['margin_low']
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
X_missing = missing_values[features]
predictions = model.predict(X_missing)
```

```python
billets.loc[billets['margin_low'].isnull(), 'margin_low'] = predictions
```

```python
billets.info()
```

```python
billets.to_csv('../Source/billets_complet.csv', sep=';', index=False)
```
